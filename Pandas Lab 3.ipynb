Open the "early_covid_by_state.csv" file

This file contains early COVID infection data from the first three-ish months of 2020. The numbers contained in each column are a running total of cases reported in that state by that date (meaning that if a state had 40 cases on day 3 and 45 cases on day 4, 5 new cases were reported on day 4). Use this data to create __graphs__ that display the following:

1. Which states had the highest rate of case growth between the first recorded date and the last (by number of cases)?
2. Which states had the highest rate of case growth (by percentage increase) during the last recorded week?
3. During the last recorded week, which states had the highest proportion of new cases nationally?

```python
# i used https://youtu.be/5rNu16O3YNE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#importing the data and indexing it by the column state
covid = pd.read_csv('early_covid_by_state.csv',index_col='State')
covid
```

```python
#seting a new var as covid 2 and creating a column that will have the values of the last day subtracted by the
#first day to show the growth. 
covid2 = covid.assign(total_days =(covid["3/9/20"] - covid["1/22/20"].astype(float)))
#only showing the values of the created column and the 50 states and making the highest values come first
covid2.sort_values(by='total_days',ascending=False).iloc[:, 50]
covid22 = covid2.sort_values(by='total_days',ascending=False).iloc[:, 50]
covid22
#NY has the most total days of growth
```

```python
#seting a new var as covid 3 and creating a column that will have the values of the last week which would be
# 2 March 2020 to 9 March 2020 by the. Then multiple by 100 
covid3 = covid.assign(last_week =(covid["3/2/20"] / covid["3/9/20"].astype(float)))*100
#only showing the values of the created column and the 50 states and making the highest values come first
covid3.sort_values(by='last_week',ascending=False).iloc[:, 50]
covid33 = covid3.sort_values(by='last_week',ascending=False).iloc[:, 50]
covid33
#WI has the highest growth percentage
```

```python
#seting a new var as covid 3 and creating a column last days that will have the values of the last day subtracted by the
#first day to show the growth. 

covid4 = covid.assign(last_days =(covid["3/9/20"] - covid["3/8/20"].astype(float)))
#only showing the values of the created column and the 50 states and making the highest values come first
covid4.sort_values(by='last_days',ascending=False).iloc[:, 50]

covid44 = covid4.sort_values(by='last_days',ascending=False).iloc[:, 50]
covid44


#NY grew by 36
```

```python
fig, (ax1, ax2) = plt.subplots (1,2)
sns.distplot(covid2, ax=ax1)
sns.boxplot(x='total_days', y='state', data = covid2.total_days, ax=ax2)
```

```python
plot = covid22.plot(kind="bar",cmap="Blues",ylim=(0,150),figsize=(14,4), color = 'red')
plot.figure.suptitle("Total Days of Covid Paitents")
plot.set_xlabel("States in the USA",fontweight="bold")
plot.set_ylabel("Cases", fontweight="bold")
```

```python
plot = covid33.plot(kind="bar",cmap="Blues",ylim=(0,150),figsize=(14,4), color = 'blue')
plot.figure.suptitle("Last Week of Covid Paitents")
plot.set_xlabel("State in USA",fontweight="bold")
plot.set_ylabel("Cases",fontweight="bold")
```
