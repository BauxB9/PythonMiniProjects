```python
def seqLn(modDict):
    #an empty array for the codons to go
    seqLns = []
    #a for loop to go through the fasta file
    for seq in modDict:
        #adds the sequence value in the array and counts them and returns the number
        seqLns.append(len(seq))
    return seqLns
    
    
    
def gcPercent(modDict):
    #an empty array for the codons to go
    gcTotal = []
    #a for loop to go through the fasta file
    for seq in modDict:
        #add the number of codon 'G' and 'C' and divide it by the len of the whole fasta
        gcTotal.append((seq.count('G') + seq.count('C')) / len(seq))
    return gcTotal
    
def vladSt(modDict):
    #an empty array for the codons to go
    seqStr = []
    #a for loop to go through the fasta file
    for seq in modDict:
        #while looping grouping them in groups of 3 and looking for ATG which is the start codon
        #i used https://www.sciencedirect.com/topics/biochemistry-genetics-and-molecular-biology/start-codon#:~:text=The%20standard%20start%20codon%20for,acids%20in%20a%20polypeptide%20chain.
        seqStr.append(seq[:3] == 'ATG')
    return seqStr
    
    
def vladSp(modDict):
    #an empty array for the codons to go
    seqStp = []
    #a for loop to go through the fasta file
    for seq in modDict:
        #while looping grouping them in groups of 3 and looking for one of the three stop codon
        #i used https://en.wikipedia.org/wiki/Stop_codon
        seqStp.append(seq[-3:] in ['TAA', 'TAG', 'TGA'])
    return seqStp
    
def main():
    #var modFile is going to read in the fasta file
        modFile = open('Mdomestica-pandasLab.fa','r')
        #var modLines is going to use the function readlines
        modLines = modFile.readlines()
        #var modDict is going to be a dictionary
        modDict = {}
        #setting var header to empty strings to accept strings from the fasta file
        header = ""
        #setting var seq to empty string to accept strings from the fasta file
        sequence = ""
        #setting var count to 0 to have a total number of seq in the fasta file
        count = 0
        #for looping the the entire fasta
        for i in range(len(modLines)):
            #looking for any 'i' within the fasta file that starts with '>'
            if modLines[i].startswith('>'):
                if sequence == "":
                    #the header for the file will take away other string characters after the '>'
                    header = modLines[i].strip()
                else:
                    #if its not like the condition above then its a seq
                    modDict[header] = sequence
                    header = modLines[i].strip()
                    sequence = ""
            #it will add each instance of the characters to the total amount of seqs
            else:
                sequence += modLines[i].strip()
                count = i
                if i == len(modLines) - 1:
                    modDict[header] = sequence
        modFile.close()
        
        
        
#these are variables that are just like the functions above with the fasta as an argument        
seqLen = seqLn(modDict)
gcTotal = gcPercent(modDict)
seqStr = vladSt(modDict)
seqStp = vladSp(modDict)
#print(modDict)
print(seqLen)
print(gcTotal)
print(seqStr)
print(seqStp)
    
#   
if __name__ == "__main__":
    main()
```
