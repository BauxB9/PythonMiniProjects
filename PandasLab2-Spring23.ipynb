```python
#i used https://youtu.be/5rNu16O3YNE
import pandas as pd

college = pd.read_csv('Most-Recent-Cohorts-Institution.csv',index_col="INSTNM", low_memory=False)
college
```

```python
loan = college[['STABBR','FTFTPCTFLOAN_POOLED_SUPP', 'MD_EARN_WNE_P10']]
loan2 = loan.sort_values('FTFTPCTFLOAN_POOLED_SUPP').dropna()
loan2.sort_values(by='FTFTPCTFLOAN_POOLED_SUPP', ascending=False)
```

```python
mdn_earnings = college[["STABBR","MD_EARN_WNE_P10","GRAD_DEBT_MDN_SUPP"]]
mdn_earnings.dropna()
```

```python
#for the total amount of nc schools
count = 0
#setting var above_7 to the total amount of undergrads above 7000
above_7000 = college.loc[college['UGDS'] > 7000]
#sorting the values by sorting students that recieved fed loans
above_7000 = above_7000.sort_values(by='PCTFLOAN', ascending=False)
# for loop for the top 50 schools
for i in range(50):
    #if the school state is in NC
    if above_7000["STABBR"].iloc[i] == "NC":
        #for the nc schools
        count = count + 1
print(count , 'from the top 50 are located in NC')
print(above_7000)
```

```python
#for the total amount of nc schools
count2 = 0 
total = 0
#resetting var above_7 to the total amount of undergrads above 7000
above_7000 = college.loc[college['UGDS'] > 7000]
#sorting the values by sorting students that recieved fed loans
above_7000 = above_7000.sort_values(by='PCTFLOAN', ascending=False)
#for loop for the top 50 schools
for i in range(50):
    #looking for the schools that distance learning
    if above_7000["DISTANCEONLY"].iloc[i] == 1:
        #var total is the amount of median earnings
        total = total + int(above_7000["MD_EARN_WNE_P10"].iloc[i])
        #the total amount of schools that are distance learning
        count2 = count2 + 1
        print (total)
print(count2 , 'from the top 50 are Distance Only')
print(above_7000)
```

```python
#setting var above_7 to the total amount of undergrads above 7000
above_7000 = college.loc[college['UGDS'] > 7000]
#sorting the values by sorting students that recieved fed loans
above_7000 = above_7000.sort_values(by='PCTFLOAN', ascending=False)
#setting the columns within GRAD_DEBT_MDN_SUPP not accept values that are privacy Suppressed and used the median amount
medianNet = above_7000[above_7000["GRAD_DEBT_MDN_SUPP"] != 'PrivacySuppressed' ]["GRAD_DEBT_MDN_SUPP"].astype(float).median()
# looking within the NC school, setting the columns within GRAD_DEBT_MDN_SUPP not accept values that are privacy Suppressed and used the median amount
medianNc = above_7000[(above_7000['STABBR']== 'NC') & (above_7000["GRAD_DEBT_MDN_SUPP"] != 'PrivacySuppressed') ]["GRAD_DEBT_MDN_SUPP"].astype(float).median()

print(medianNet)
print(medianNc)
print(above_7000)
```
