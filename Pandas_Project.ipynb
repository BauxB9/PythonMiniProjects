```python
#i used https://youtu.be/5rNu16O3YNE
#i used https://github.com/vtyeh/pandas-challenge to help me understand how to structure it and ranking
#i followed the example that were in class with my peers using functions to create the listing

import pandas as pd
import numpy as np

#ranking colleges based off of these websites below

#https://www.usnews.com/education/best-colleges/articles/how-us-news-calculated-the-rankings
#https://www.forbes.com/sites/michaeltnietzel/2022/08/30/americas-military-service-academies-an-outstanding-option-for-undergraduate-education/?sh=53e88c196b85
#https://www.forbes.com/sites/emmawhitford/2022/06/09/the-strongest-and-weakest-colleges-in-america---behind-forbes-2022-financial-grades/?sh=3ab89b47ffad

#College Score Card possible factors
#acceptance rate: ADM_RATE, median earing: MD_EARN_WNE_P10, retention rate: RET_FT4, grad rates: C150_4, loans: GRAD_DEBT_MDN_SUPP, fullpaid loans: BBRR2_FED_UG_PAIDINFULL_SUPP
#graduation: C150_4_POOLED_SUPP, student return: RET_FT4_POOLED_SUP, AVGFACSAL, TUITFTE, ADM_RATE_ALL
#grad rates:'C150_4', retention rate:'RET_FT4', median earing:'MD_EARN_WNE_P10', student_fac: 'STUFACR', Admin rate: 'ADM_RATE'

#Readinf the data in and converting all PrivacySuppressed into Nan, I am using .dropna() for the after my function
colleges = pd.read_csv("Most-Recent-Cohorts-Institution.csv",index_col="INSTNM", low_memory=False)
#replacing the PS to NaN
colleges2 = colleges.replace('PrivacySuppressed', np.nan)
#dropping the NaN values
colleges2 = colleges2.dropna(subset=['ADM_RATE','RET_FT4','C150_4','MD_EARN_WNE_P10','GRAD_DEBT_MDN_SUPP','C150_4_POOLED_SUPP', 'TUITFTE', 'AVGFACSAL'])
#colleges2 = colleges2.dropna(subset=['C150_4_POOLED_SUPP', 'RET_FT4_POOLED_SUP', 'AVGFACSAL', 'TUITFTE', 'ADM_RATE_ALL'])



#function for the factors that are choosen
def scoreCard(row):
    #conveting admin rates into a floating (decimal accepting numbers) for var admin
    admin = float(row['ADM_RATE_ALL'])
    #converting retention into a float for var rten
    rten = float(row['RET_FT4']) 
    #converting total graduations into a float for var gard
    grad = float(row['C150_4_POOLED_SUPP'])
    #converting earnigns into a float var earnings
    earnings = float(row['MD_EARN_WNE_P10'])
    #converting debt into a float var debt
    debt = float(row['GRAD_DEBT_MDN_SUPP'])
    #converting current graduations to a float var grad2
    grad2 = float(row['C150_4'])
    #converting student tution help into a float var stu
    stu = float(row['TUITFTE'])
    #converting student fincancal help into a float var stu2
    stu2 = float(row['AVGFACSAL'])
    
    
    #creating the percentages for the values out of .1 (100%)
    uniPercent = {
        'admin': .2,
        'rten' : .2,
        'earnings' : .4,
        'debt' : .05,
        'stu2' : .05,
        'grad' : .1
    }
    
    #scoring system
    uniScore = (
        #using the var uniPercent and multiplying it to itself to make the score a smaller number and adding them up
        uniPercent['admin'] * admin 
        + uniPercent['rten'] * rten 
        + uniPercent['stu2'] * stu2 
        + uniPercent['grad'] * grad
        #below is the going to be the ROI (return of investment) for going to the schools 
        #schools with smaller earnings and higher debt will be way at the bottom 
        / ( uniPercent['earnings'] * earnings 
        - uniPercent['debt'] * debt)
    
    )
   
    #returning the scoring points 
    return uniScore

#making sure the data within the file uses the scoreCard function to tally the points
colleges2['uniScore'] = colleges2.apply(scoreCard, axis = 1)

#creating a var college_score to sore the values from the the file
college_score = colleges2.sort_values(by = 'uniScore', ascending=False)

#var college score are sorted by the scoring system with the top 10 schools
college_score[['uniScore', 'STABBR']].head(10)


```

```python
#https://www.forbes.com/video/6311598410112/the-top-5-colleges-in-the-us/?sh=46787428f2ab
#https://www.forbes.com/pictures/63062f99a431a1d1ebc1d377/top-25-lowest-debt/?sh=4cb895ae261a
#https://www.forbes.com/pictures/63067b2c97683f2a951b5fd4/1-harvard-university/?sh=1e2616cd43f3
#chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://collegescorecard.ed.gov/assets/InstitutionDataDocumentation.pdf

#College Score Card possible factors:
#state: "STABBR","median earing:'MD_EARN_WNE_P10',"GRAD_DEBT_MDN_SUPP",grad rates:'C150_4', retention rate:'RET_FT4'
#acceptance rate: ADM_RATE, median earing: MD_EARN_WNE_P10, retention rate: RET_FT4, grad rates: C150_4, loans: GRAD_DEBT_MDN_SUPP, fullpaid loans: BBRR2_FED_UG_PAIDINFULL_SUPP
#graduation: C150_4_POOLED_SUPP, student return: RET_FT4_POOLED_SUP, AVGFACSAL, TUITFTE, ADM_RATE_ALL
#grad rates:'C150_4', retention rate:'RET_FT4', median earing:'MD_EARN_WNE_P10', student_fac: 'STUFACR', Admin rate: 'ADM_RATE'



#Readind the data in and converting all PrivacySuppressed into Nan, I am using .dropna() for the after my function
mySchool = pd.read_csv("Most-Recent-Cohorts-Institution.csv", low_memory=False)
#replacing the PS to NaN
mySchool = mySchool.replace('PrivacySuppressed', np.nan)
#Dropping NaN values
mySchool = mySchool.dropna(subset=['ADM_RATE','RET_FT4','BBRR2_FED_UG_PAIDINFULL_SUPP','MD_EARN_WNE_P10'
                                   ,'GRAD_DEBT_MDN_SUPP', 'VETERAN', 'PAR_ED_PCT_1STGEN', 'BBRR2_FED_UG_PAIDINFULL_SUPP',
                                  'TUITFTE'])



#function for the factors that are choosen
def scoreCard2(row):
    #conveting admin rates into a floating (decimal accepting numbers) for var admin
    admin = float(row['ADM_RATE_ALL'])
    #converting retention into a float for var rten
    rten = float(row['RET_FT4']) 
    #converting total graduations into a float for var gard
    grad = float(row['C150_4_POOLED_SUPP'])
    #converting earnigns into a float var earnings
    earnings = float(row['MD_EARN_WNE_P10'])
    #converting debt into a float var debt
    debt = float(row['GRAD_DEBT_MDN_SUPP'])
    #converting current graduations to a float var grad2
    grant = float(row['BBRR2_FED_UG_PAIDINFULL_SUPP'])
    #converting student veteran population help into a float var stu
    stu = float(row['VETERAN'])
    #converting  first-generation on the FAFSA into a float var stu2
    stu2 = float(row['PAR_ED_PCT_1STGEN'])
    #converting student fully repaid debt into a float var stu
    debt2 = float(row['BBRR2_FED_UG_PAIDINFULL_SUPP'])
    #converting  first-generation on the FAFSA into a float var stu2
    stu3 = float(row['TUITFTE'])
    
    
    #creating the percentages for the values out of 1 (100%)
    uniPercent = {
        'admin': .1,
        'rten' : .1,
        'earnings' : .1,
        'debt' : .1,
        'stu2' : .1,
        'grad' : .1,
        'grant' : .1,
        'stu' : .1,
        'debt2' : .1,
        'stu3' : .1
    }
    
    #scoring system
    uniScore2 = (
        uniPercent['admin'] * admin
        + uniPercent['rten'] * rten
        + uniPercent['stu2'] * stu2
        + uniPercent['stu'] * stu
        + uniPercent['stu3'] * stu3
        + uniPercent['grad'] * grad) 
        + 
        #The ROI based on your earning potential and the amount of money/grants you can get
        #with amount of debt you can get subtracted by the ability to pay it off within 8 years
        (uniPercent['earnings'] * earnings + uniPercent['grant'] * grant
        - (uniPercent['debt'] * debt - uniPercent['debt2'] * debt2))

    return uniScore2

#making sure the data within the file uses the scoreCard function to tally the points
mySchool['uniScore2'] = mySchool.apply(scoreCard2, axis = 1)

#creating a var my_score to sort the values from the the file
my_score = mySchool.sort_values(by = 'uniScore2', ascending=False)

#sorting the data and giving a ranking
my_score['Rank'] = my_score['uniScore2'].rank(ascending=False)

#var tamu scored the sorted by the scoring system with the 4 columns
tamu = my_score[['Rank','uniScore2','INSTNM','STABBR']]


#I had to unindex the column to be able to look for the school
tamu[tamu['INSTNM']== 'Texas A & M University-College Station']
```
